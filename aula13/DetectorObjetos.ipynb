```python
import cv2
import matplotlib.pyplot as plt

class FaceDetector:
    def __init__(self, faceCascadePath):
        self.faceCascade = cv2.CascadeClassifier(faceCascadePath)
        
    def detect(self, image, scaleFactor = 1.01, minNeighbors = 5, minSize = (50, 50)):
        rects = self.faceCascade.detectMultiScale(image, scaleFactor = scaleFactor, minNeighbors = minNeighbors, minSize = minSize, flags = cv2.CASCADE_SCALE_IMAGE)
        return rects
    
image = cv2.imread('a5.bmp')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
#fd = FaceDetector('haarcascade_frontalface_default.xml')
fd = FaceDetector('myhaar.xml')
faceRects = fd.detect(gray, scaleFactor = 1.001, minNeighbors = 3, minSize = (25, 25))
print("I found {} face(s)".format(len(faceRects)))

for (x, y, w, h) in faceRects:
    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)

plt.imshow(image)
```
